# Analizando una Imagen

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread("beagle.jpg")

print("Tipo de objeto:", type(img))
print("Número de dimensiones:", img.ndim)
print("Forma de la imagen:", img.shape)
print("Cantidad total de valores:", img.size)
print("Tipo de datos:", img.dtype)
print("Píxel localizado en [10, 10]:", img[10, 10])
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.title("Imagen original del beagle")
plt.axis("off")
plt.show()
```

```python
promedio_rgb = img.mean(axis=(0, 1))

print("Promedio del canal rojo:", promedio_rgb[0])
print("Promedio del canal verde:", promedio_rgb[1])
print("Promedio del canal azul:", promedio_rgb[2])
```

```python
imagen_gris = (
    0.299 * img[:, :, 0]
    + 0.587 * img[:, :, 1]
    + 0.114 * img[:, :, 2]
)

plt.figure(figsize=(8, 6))
plt.imshow(imagen_gris, cmap="gray")
plt.title("Beagle en escala de grises")
plt.axis("off")
plt.show()
```

```python
img[:, :, 0]  # canal rojo
img[:, :, 1]  # canal verde
img[:, :, 2]  # canal azul
```

```python
alto, ancho = img.shape[:2]

recorte = img[
    alto // 4 : 3 * alto // 4,
    ancho // 4 : 3 * ancho // 4
]

plt.figure(figsize=(6, 6))
plt.imshow(recorte)
plt.title("Recorte central de la imagen")
plt.axis("off")
plt.show()

print("Forma original:", img.shape)
print("Forma del recorte:", recorte.shape)
```

```python
imagen_volteada = img[:, ::-1, :]

plt.figure(figsize=(8, 6))
plt.imshow(imagen_volteada)
plt.title("Imagen volteada horizontalmente")
plt.axis("off")
plt.show()
```

### Conclusión

Una imagen digital puede representarse como un arreglo tridimensional. Cada píxel contiene valores correspondientes a los canales rojo, verde y azul. Con NumPy analicé la estructura de la imagen, consulté un píxel, calculé los colores promedio, convertí la imagen a escala de grises, realicé un recorte y la volteé horizontalmente.
